```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Time the bus will arrive ...

t_1 = np.sort(np.append(
    np.random.uniform(0, 4.0, 35),
    np.random.uniform(8.0, 11.0, 35),
))  # The input coordinates must be sorted
print(t_1)
A = 20.0
y_1 = np.random.randn(len(t_1)) + A * np.sin(2.0 * np.pi * t_1 / 7)
yerr = np.ones(len(t_1)) * 1.0
y_1 += np.random.randn(len(t_1)) * yerr 
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.errorbar(t_1, y_1, yerr=yerr, ls='none')
```

```python
import celerite
from celerite import terms

# A periodic component
Q = 1.0
w0 = 7.0
S0 = np.var(y_1) / (w0 * Q)
bounds = dict(log_S0=(-25, 15), log_Q=(-15, 25), log_omega0=(-15, 15))
kernel = terms.SHOTerm(log_S0=np.log(S0), log_Q=np.log(Q), log_omega0=np.log(w0),
                        bounds=bounds)
gp = celerite.GP(kernel, mean=0, fit_mean=False)
gp.compute(t_1, yerr)  # You always need to call compute once.
print("Initial log likelihood: {0}".format(gp.log_likelihood(y_1)))
```

```python
def log_probability(params):
    gp.set_parameter_vector(params)
    lp = gp.log_prior()
    if not np.isfinite(lp):
        return -np.inf
    #if params[0] < 0:
    #    return -np.inf
    return gp.log_likelihood(y_1) + lp
```

```python
import emcee

initial = gp.get_parameter_vector()
ndim, nwalkers = len(initial), 32
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability)

print("Running burn-in...")
p0 = initial + 1e-8 * np.random.randn(nwalkers, ndim)
p0, lp, _ = sampler.run_mcmc(p0, 6000)

print("Running production...")
sampler.reset()
sampler.run_mcmc(p0, 2000);
```

```python
import corner
corner.corner(sampler.flatchain)
plt.show()
```

```python
# Plot the data.
plt.errorbar(t_1, y_1, yerr=yerr, fmt=".k", capsize=0, label='Data')
t_2 = np.arange(0, 30, 0.1)
# Plot 24 posterior samples.
samples = sampler.flatchain
for s in samples[np.random.randint(len(samples), size=53*20)]:
    gp.set_parameter_vector(s)
    mu = gp.predict(y_1, t_2, return_cov=False)
    a = plt.plot(t_2, mu, color='r', alpha=0.3)

plt.ylabel(r"$y$")
plt.xlabel(r"$t$")
plt.legend()
plt.title("posterior predictions");
```
